```python
import numpy as np
import quandl
import pandas as pd
import matplotlib.pyplot as plt
quandl.ApiConfig.api_key = 'k5hpQXTQmkr3f2cBYTDb'
```

```python
names_of_company = ['HBL']
HBL = pd.DataFrame()
for names in names_of_company:
    x = quandl.get('PSX/%s' %names, start_date='2011-1-1', end_date='2021-7-30')
    x['company'] = names
    HBL = HBL.append(x)
    
HBL = HBL.reset_index
HBL = HBL('Date')
HBL_Turnover = HBL['Turnover']
HBL_date = HBL['Date']
HBL_daily = HBL['Open'] - HBL['Close'].shift(1)
HBL_monthMean = HBL.resample('M', on='Date').mean()
HBL_yearMean = HBL.resample('Y', on='Date').mean()
HBL_dailyReturn = (HBL['Close'].shift(1) / HBL['Close']) - 1
HBL
```

```python
names_of_company = ['ABL']
ABL = pd.DataFrame()
for names in names_of_company:
    x = quandl.get('PSX/%s' %names, start_date='2011-1-1', end_date='2021-7-30')
    x['company'] = names
    ABL = ABL.append(x)
    
ABL = ABL.reset_index
ABL = ABL('Date')
ABLs_Turnover = ABL['Turnover']
ABLs_date = ABL['Date']
ABLs_daily = ABL['Open'] - ABL['Close'].shift(1)
ABLs_monthMean = ABL.resample('M', on='Date').mean()
ABLs_yearMean = ABL.resample('Y', on='Date').mean()
ABLs_dailyReturn = (ABL['Close'].shift(1) / ABL['Close']) - 1
ABL
```

```python
names_of_company = ['UBL']
UBL = pd.DataFrame()
for names in names_of_company:
    x = quandl.get('PSX/%s' %names, start_date='2011-1-1', end_date='2021-7-30')
    x['company'] = names
    UBL = UBL.append(x)
    
    
UBL = UBL.reset_index
UBL = UBL('Date')
UBL_Turnover = UBL['Turnover']
UBL_date = UBL['Date']
UBL_daily= UBL['Open'] - UBL['Close'].shift(1)
UBL_monthMean = UBL.resample('M', on='Date').mean()
UBL_yearMean = UBL.resample('Y', on='Date').mean()
UBL_dailyReturn = (UBL['Close'].shift(1) / UBL['Close']) - 1
UBL
```

```python
names_of_company = ['BOP']
BOP = pd.DataFrame()
for names in names_of_company:
    x = quandl.get('PSX/%s' %names, start_date='2011-1-1', end_date='2021-7-30')
    x['company'] = names
    BOP = BOP.append(x)
BOP = BOP.reset_index
BOP = BOP('Date')
BOP_Turnover = BOP['Turnover']
BOP_date = BOP['Date']
BOP_daily = BOP['Open'] - BOP['Close'].shift(1)
BOP_monthMean = BOP.resample('M', on='Date').mean()
BOP_yearMean = BOP.resample('Y', on='Date').mean()
BOP_dailyReturn = (BOP['Close'].shift(1) / BOP['Close']) - 1
BOP
```
